# Data Visualization

```python
import matplotlib.pyplot as plt
```

# Bargraph

```python
sample_df = df.head(10)
plt.bar(sample_df["model"], sample_df["range_km"], color="skyblue")

plt.title("Range of EV Models")
plt.xlabel("Model")
plt.ylabel("Range (km)")
plt.xticks(rotation=90)  
plt.grid(axis='y', alpha=0.6)
plt.show()
```

## Scatterplot
- alpha=x for trasparency
- c for color of scatterplot and colorbar for showing how the colors correspond to the numeric values
- s=x for size
- edgecolors = "color"
- linewidths = x
- label = "String"

```python
plt.scatter(df["top_speed_kmh"], df["range_km"])
plt.title("Top Speed vs Range")
plt.xlabel("Top Speed (km/h)")
plt.ylabel("Range (km)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(
    df["top_speed_kmh"], 
    df["range_km"], 
    c=df["acceleration_0_100_s"],  # Color by acceleration time
)  
plt.colorbar(label="Acceleration 0-100 km/h (s)")  # Add color comparision
plt.title("Top Speed vs Range colored by Acceleration")
plt.xlabel("Top Speed (km/h)")
plt.ylabel("Range (km)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(
    df["top_speed_kmh"], 
    df["range_km"],
    alpha=1,  # Transparency
    edgecolors='red', #Edge Color
    linewidths = 1, #Line that Surrownds
    s = 50,
    marker="D"    # Size 
    )
plt.title("Top Speed vs Range colored by Acceleration")
plt.xlabel("Top Speed (km/h)")
plt.ylabel("Range (km)")
plt.grid(True)
plt.show()
```

```python
df2 = df.head(12)
plt.plot(df2['top_speed_kmh'], marker='o')
plt.title('Top Speed of Electric Cars')
plt.ylabel('Top Speed (km/h)')
plt.grid(True)
plt.show()
```

```python
plt.hist(df['top_speed_kmh'], bins=10, edgecolor='black')
plt.title('Distribution of Top Speed (km/h)')
plt.xlabel('Top Speed (km/h)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df['battery_capacity_kWh'].dropna())
plt.title('Distribution of Battery Capacity (kWh)')
plt.ylabel('Battery Capacity (kWh)')
plt.grid(True)
plt.show()
```
